```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (Js)
m = 9.11e-31  # Mass of electron (kg)
L = 1e-9  # Length of the box (in meters, e.g., 1 nm)
num_points = 1000  # Number of spatial points
x = np.linspace(0, L, num_points)  # Discretized positions inside the box

# Eigenvalue calculation (energy levels)
def energy_level(n):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m * L**2)

# Solve the Schrdinger equation numerically for different n
def wavefunction(n, x):
    # The general solution to the Schrdinger equation for particle in a box
    # is _n(x) = sqrt(2/L) * sin(n * pi * x / L)
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

# Plot the first few wavefunctions and energy levels
plt.figure(figsize=(10, 6))
n_values = [1, 2, 3, 4]  # The first few energy levels (n = 1, 2, 3, 4)

for n in n_values:
    psi_n = wavefunction(n, x)
    E_n = energy_level(n)
    
    plt.plot(x, psi_n, label=f"n={n}, E={E_n:.3e} J")

# Adding plot details
plt.title("Wavefunctions for Particle in a 1D Box")
plt.xlabel("Position (m)")
plt.ylabel("Wavefunction (x)")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
